```python
#Librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Imágen a Escala de grises

```python
# Imagen en escala de grises
image = cv2.imread('ave.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(12, 5))
plt.imshow(gray, cmap='gray')
plt.title('Imagen en escala de grises')
plt.axis('off')
```

**Calcular histograma con "cv2.calcHist"**



```python
# Calcular histograma con cv2.calcHist()
hist_cv2 = cv2.calcHist([gray], [0], None, [256], [0, 256])

plt.plot(hist_cv2, color='black')
plt.title('Histograma con cv2.calcHist()')
plt.xlabel('Valor de intensidad')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

**Calcular histograma con "plt.hist"**

```python
#plt.hist():
plt.figure()
plt.hist(gray.ravel(), bins=256, range=(0, 256), color='gray')
plt.title('Histograma con plt.hist()')
plt.xlabel('Valor de intensidad')
plt.ylabel('Frecuencia')
plt.show()
```
