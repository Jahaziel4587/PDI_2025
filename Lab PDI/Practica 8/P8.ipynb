```
import numpy as np
import matplotlib.pyplot as plt
import cv2
#Carga de imágenes
x,y=(600,600)
img=cv2.imread('radiografia_1.jpg',0)
img=cv2.resize(img,(x,y))
#conversión
img_32=img.astype(np.float32)
four=np.fft.fft2(img_32)
frr=np.fft.fftshift(four)

frr_abs=np.abs(frr)

frr_log=20*np.log10(frr_abs)

img_frr=np.uint8(255*frr_log/np.max(frr_log))

F1=np.arange(-x/2+1,x/2+1,1)
F2=np.arange(-y/2+1,y/2+1,1)
[X,Y]=np.meshgrid(F1,F2)
D=np.sqrt(X*2+Y*2)
D=D/np.max(D)

Do=.20

huv=np.zeros((x,y))

for i in range(x):
    for j in range(y):
        if(D[i,j]<Do):
            huv[i,j]=1

huv=1-huv

guv=huv*frr

guv_abs=np.abs(guv)
guv_abs=np.uint8(255*guv_abs/np.max(guv_abs))

gxy=np.fft.ifft2(guv)
gxy=np.abs(gxy)
gxy=np.uint8(gxy)

plt.figure(figsize=(16,12))

plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(img_frr,cmap='gray')
plt.title('Espectro Fourier Logarítmica')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(np.uint8(255*huv),cmap='gray')
plt.title(f'Filtro paso alto (A. Corte = {Do*100}%)')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(gxy,cmap='gray')
plt.title('Imagen filtrada')
plt.axis('off')

plt.tight_layout()
plt.show()
```
