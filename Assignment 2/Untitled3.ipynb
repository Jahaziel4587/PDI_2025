```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_gray = cv2.imread("Color.jpg", cv2.IMREAD_GRAYSCALE)
brillo = 50
img_brillo = cv2.add(img_gray, brillo)
contraste = 1.5
img_contraste = cv2.convertScaleAbs(img_gray, alpha=contraste, beta=0)  # Alpha -> Contraste, Beta -> Brillo

plt.figure(figsize=(10,5))
plt.subplot(1,3,1), plt.imshow(img_gray, cmap="gray"), plt.title("Original")
plt.subplot(1,3,2), plt.imshow(img_brillo, cmap="gray"), plt.title("Brillo ")
plt.subplot(1,3,3), plt.imshow(img_contraste, cmap="gray"), plt.title("Contraste ")
plt.show()
```

```python
# Ecualizacin de histograma
img_eq = cv2.equalizeHist(img_gray)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1), plt.imshow(img_gray, cmap="gray"), plt.title("Original")
plt.subplot(1,2,2), plt.imshow(img_eq, cmap="gray"), plt.title("Ecualizacin de Histograma")
plt.show()
```

```python
#OPERACIN AND
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread("Color.jpg", cv2.IMREAD_GRAYSCALE)

#Mscara binaria
mask = np.zeros(img.shape, dtype=np.uint8)
cv2.circle(mask, (img.shape[1]//2, img.shape[0]//2), 100, 255, -1)  # Crculo blanco

# Aplicar AND
img_and = cv2.bitwise_and(img, mask)

#imgenes:
plt.figure(figsize=(10,5))
plt.subplot(1,3,1), plt.imshow(img, cmap="gray"), plt.title("Imagen Original")
plt.subplot(1,3,2), plt.imshow(mask, cmap="gray"), plt.title("Mscara")
plt.subplot(1,3,3), plt.imshow(img_and, cmap="gray"), plt.title("AND (Interseccin)")
plt.show()
```

```python
#OPERACION OR
img_or = cv2.bitwise_or(img, mask)

#Imgenes:
plt.figure(figsize=(10,5))
plt.subplot(1,3,1), plt.imshow(img, cmap="gray"), plt.title("Imagen Original")
plt.subplot(1,3,2), plt.imshow(mask, cmap="gray"), plt.title("Mscara")
plt.subplot(1,3,3), plt.imshow(img_or, cmap="gray"), plt.title("OR (Unin)")
plt.show()
```

```python
#OPERACIN XOR

img_xor = cv2.bitwise_xor(img, mask)

#Imgenes:
plt.figure(figsize=(10,5))
plt.subplot(1,3,1), plt.imshow(img, cmap="gray"), plt.title("Imagen Original")
plt.subplot(1,3,2), plt.imshow(mask, cmap="gray"), plt.title("Mscara")
plt.subplot(1,3,3), plt.imshow(img_xor, cmap="gray"), plt.title("XOR (Diferencia)")
plt.show()
```

```python
#OPERACIN NOT
img_not = cv2.bitwise_not(img)

#Imgenes:
plt.figure(figsize=(10,5))
plt.subplot(1,2,1), plt.imshow(img, cmap="gray"), plt.title("Imagen Original")
plt.subplot(1,2,2), plt.imshow(img_not, cmap="gray"), plt.title("NOT (Inversin)")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img_gray = cv2.imread("Color.jpg", cv2.IMREAD_GRAYSCALE)

# Aplicar NOT antes de la umbralizacin
img_gray_not = cv2.bitwise_not(img_gray)

# Aplicar umbralizacin inversa
valor_umbral = 180
_, img_thresh_inv = cv2.threshold(img_gray_not, valor_umbral, 255, cv2.THRESH_BINARY_INV)

#imgenes:
plt.figure(figsize=(10,5))
plt.subplot(1,3,1), plt.imshow(img_gray, cmap="gray"), plt.title("Escala de Grises")
plt.subplot(1,3,2), plt.imshow(img_gray_not, cmap="gray"), plt.title("Aplicacin de NOT")
plt.subplot(1,3,3), plt.imshow(img_thresh_inv, cmap="gray"), plt.title("Umbral Inverso")
plt.show()
```

```python
#imagen en color
img_color = cv2.imread("Color.jpg")

#canales B, G, R
b, g, r = cv2.split(img_color)

#NOT a cada canal
b_not = cv2.bitwise_not(b)
g_not = cv2.bitwise_not(g)
r_not = cv2.bitwise_not(r)

#canales
plt.figure(figsize=(10,5))
plt.subplot(2,2,1), plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)), plt.title("Imagen Original")
plt.subplot(2,2,2), plt.imshow(b_not, cmap="gray"), plt.title("Canal Azul (B) - NOT")
plt.subplot(2,2,3), plt.imshow(g_not, cmap="gray"), plt.title("Canal Verde (G) - NOT")
plt.subplot(2,2,4), plt.imshow(r_not, cmap="gray"), plt.title("Canal Rojo (R) - NOT")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#imagen en escala de grises
img_gray = cv2.imread("Color.jpg", cv2.IMREAD_GRAYSCALE)

#umbralizacin inversa
valor_umbral = 127
_, img_bin = cv2.threshold(img_gray, valor_umbral, 255, cv2.THRESH_BINARY_INV)

#rea del objeto segmentado
area_objeto = np.count_nonzero(img_bin)

print(f"rea del objeto segmentado: {area_objeto} pxeles")
```

```python
#bounding box
x, y, w, h = cv2.boundingRect(img_bin)

print(f"Bounding Box - x: {x}, y: {y}, ancho: {w}, alto: {h}")
```

```python
#imagen en color
img_color = cv2.imread("Color.jpg")

#bounding box en azul
cv2.rectangle(img_color, (x, y), (x + w, y + h), (255, 0, 0), 2)

#imagen con el bounding box:
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title("Bounding Box sobre la Imagen")
plt.axis("off")
plt.show()
```
